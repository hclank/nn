```python
class Value:
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self._prev = set(_children)
        self.grad = 0
        self._op = _op
        self.label = label

    def __repr__(self):
        return f"Value(data={self.data})"

    def __add__(self, other):
        out  = Value(self.data + other.data, (self, other), _op='+')
        return out

    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), _op='*')
        return out

a = Value(2.0, label='a')
b = Value(-3.0, label='b')
c = Value(5.0, label='c')

d = a*b+c
```
